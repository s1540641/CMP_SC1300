```python
salary = 0.0
total = 0
counter = 0

while salary >= 0:
    salary = float(input("Enter a valid salary value (>=0), -1 to exit:"))
    if salary > 0:
        total = total + salary
        counter = counter + 1
        
print("Total of the salaries: ", total)

if counter > 0: # make sure user entered at least 1 valid salary value
    average = total/counter
    print("Average of the salaries: ", average)
```

```python
salary = 0.0
total = 0
counter = 0

while salary != -1:
    salary = float(input("Enter a valid salary value (>=0), -1 to exit:"))
    if salary > 0:
        total = total + salary
        counter = counter + 1

print("Total of " + str(counter) + " salaries: " + str(total))
print("Total of ", counter, "salaries: ", total)
#print("Total of %d salaries: %d", %(counter,total))

if counter > 0: # make sure user entered at least 1 valid salary value
    average = total/counter
    print("Average of the salaries: ", average)
```

```python
total = 0
counter = 0
count_passed = 0
count_failed = 0

grade = int(input("Enter a grade (0 to 100), -1 to exit: "))
while grade != -1:
    total = total + grade
    counter = counter + 1

    if grade >= 70:
        count_passed = count_passed + 1
        print("Passed")
    else:
        count_failed = count_failed + 1
        print("Failed")
    
    grade = int(input("Enter a grade (0 to 100), -1 to exit: "))

print(total)
if counter != 0:
    print(total/counter)

print("Number of people passed: ", count_passed)
print("Number of people failed: ", count_failed)
```

```python
total = 0.0
count = 0
inputStr = input("Enter value: ")
while inputStr != "" :
   value = float(inputStr)
   total = total + value
   count = count + 1
   inputStr = input("Enter value: ")

if count > 0 : 
   average = total / count
else :
   average = 0.0

print(average)
```

```python
count_As = 0
count_Bs = 0
count_Cs = 0
count_Ds = 0
count_Fs = 0
numOfStudents = 0

grade = int(input("Enter student grade (1-100), -1 if you are done with entering grades: "))
while grade != -1 :
    numOfStudents = numOfStudents + 1
    if grade >= 90 :
        letterGrade = "A"
        count_As = count_As + 1
    elif grade >= 80 and grade < 90 :
        letterGrade = "B"
        count_Bs = count_Bs + 1
    elif grade >= 70 and grade < 80 :
        letterGrade = "C"
        count_Cs = count_Cs + 1
    elif grade >= 60 and grade < 70 :
        letterGrade = "D"
        count_Ds = count_Ds + 1
    else :
        count_Fs = count_Fs + 1
        letterGrade = "F"
    print(letterGrade)
    grade = int(input("Enter student grade (1-100), -1 if you are done with entering grades: "))

print("Number of students ", numOfStudents)
print("Number of As ", count_As)
print("Number of Bs ", count_Bs)
print("Number of Cs ", count_Cs)
print("Number of Ds ", count_Ds)
print("Number of Fs ", count_Fs)

```
