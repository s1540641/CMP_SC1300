```python
total = 0
counter = 0

numOfStudents = int(input("Enter number of students: "))
i = 0
while i < numOfStudents: # n students, i will get values of 0,...n-1
    grade = int(input("Enter a grade for the student (0 to 100), -1 to exit: "))
    while grade != -1:
        total = total + grade
        counter = counter + 1
        grade = int(input("Enter a grade for the student (0 to 100), -1 to exit: "))
        
    print(total)
    if counter != 0:
        print(total/counter)
    i = i + 1
```
