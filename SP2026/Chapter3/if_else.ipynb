```python
number = 10

if number > 10:
    print("Number %d is greater than 10" %(number))
elif number < 10:
    print("Number %d is less than 10" %(number))
else: 
    print("Number %d is equal to 10" %(number))
```

```python
number = 10

if number > 10:
    print("Number %d is greater than 10" %(number))
else:
    print("Number %d is less than or equal to 10" %(number))

print("Number %d is greater than 10" %(number)) if number > 10 else print("Number %d is less than or equal to 10" %(number))
```

```python
number1 = int(input("Number 1: "))
number2 = int(input("Number 2: "))

print("number1 is", number1)
print("number2 is", number2)

if number1 == number2:
    print("number1 is equal to number2")
if number1 != number2:
    print("number1 is not equal to number2")
if number1 > number2:
    print("number1 is greater than number2")
if number1 < number2:
    print("number1 is smaller than number2")
```
