```python
text = "aba12ga9"
text2 = "12345"
text3 = "abcde"
text4 = ""

print(text.isalnum()) # letters or digits
print(text.isalpha()) # only letters
print(text.isdigit()) # only digits
print(text2.isdigit()) # only digits
print(text3.isalpha()) # only digits

print(text4.isalpha())
```

```python
text1 = "ALL UPPERCASE"
text2 = "A mix Of Up and Lower"
text3 = "all lowercase"
text4 = "   \n \t"

print(text1.islower())
print(text2.islower())
print(text3.islower())
print(text4.isspace())

print((text3.upper()).islower())
print((text3.upper()).isupper())
```
