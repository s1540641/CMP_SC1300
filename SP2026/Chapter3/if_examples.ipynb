```python
# this program will get a grade from 0 to 100 from the user
# and print the letter grade corresponding to that grade
grade = int(input("Enter the grade (0 to 100): "))

if grade >= 90:
    print("Letter grade is A")
elif grade >= 80:
    print("Letter grade is B")
elif grade >= 70:
    print("Letter grade is C")
elif grade >= 60:
    print("Letter grade is D")
else:
    print("Letter grade is F")

print("Outside of the if elif else compound block")
```
